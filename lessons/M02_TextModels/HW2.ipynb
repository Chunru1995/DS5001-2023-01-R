```python
text_file = f"{data_home}/gutenberg/pg161.txt"
csv_file = f"{data_home}/output/austen_sense.csv" #file we will create
```

```python
OHCO = ['chap_num', 'para_num', 'sent_num', 'token_num']
```

```python
LINES = pd.DataFrame(open(text_file, 'r', encoding='utf-8-sig').readlines(), columns=['line_str'])
LINES.index.name='line_num'
LINES.line_str=LINES.line_str.str.replace(r'\n+', ' ', regex=True).str.strip() #stripping out new lines
```

```python
title = LINES.loc[0].line_str.replace('The Project Gutenberg EBook of', '')
```

```python
clip_pats = [
    r"\*\*\* START OF THIS PROJECT", 
    r"\*\*\* END OF THIS PROJECT"
] #takes boiler plate of gutenberg out
```

```python
pat_a = LINES.line_str.str.match(clip_pats[0]) 
pat_b = LINES.line_str.str.match(clip_pats[1]) 
```

```python
pat_a.sum(), pat_b.sum() 
```

```python
line_a = LINES.loc[pat_a].index[0]+1 #first line after end of gberg text
line_b = LINES.loc[pat_b].index[0]-1 #line before beginning of gberg text
```

```python
line_a, line_b
```

```python
LINES = LINES.loc[line_a : line_b]
```

```python
LINES.head(10)
```

```python
LINES.tail(10)
```

```python
chap_pat = r"^\s*(?:chapter|letter)\s+\d+" #identify headers for chapters
```

```python
chap_lines = LINES.line_str.str.match(chap_pat, case=False) #returns a truth vector
```

```python
LINES.loc[chap_lines] #use as filter for dataframe
```

```python
LINES.loc[chap_lines, 'chap_num'] = [i+1 for i in range(LINES.loc[chap_lines].shape[0])] #new column for chapter headers
```

```python
LINES.loc[chap_lines]
```

```python
LINES.chap_num = LINES.chap_num.ffill() #each line associated with a chapter in this new column
```

```python
LINES.sample(10)
```

```python
LINES = LINES.dropna(subset=['chap_num']) #remove everything before ch1
#LINES = LINES.loc[~LINES.chap_num.isna()] #remove everything before ch1 alt method
LINES = LINES.loc[~chap_lines] #remove chapter heading lines: their work is done
LINES.chap_num = LINES.chap_num.astype('int') #convert chap_num from float to int value
```

```python
LINES.sample(10)
```

```python
OHCO[:1]
```

```python
CHAPS = LINES.groupby(OHCO[:1]).line_str.apply(lambda x: '\n'.join(x)).to_frame('chap_str') #make big string for each chapter
```

```python
CHAPS.head(10)
```

```python
CHAPS['chap_str'] = CHAPS.chap_str.str.strip() #removes white space before / after each chapter
```

```python
CHAPS.head(10)
```

```python
para_pat = r'\n\n+' 
```

```python
PARAS = CHAPS['chap_str'].str.split(para_pat, expand=True).stack()\
    .to_frame('para_str').sort_index() #good practice to sort index
PARAS.index.names= OHCO[:2]
```

```python
PARAS.head()
```

```python
PARAS['para_str'] = PARAS['para_str'].str.replace(r'\n', ' ', regex=True)
PARAS['para_str'] = PARAS['para_str'].str.strip()
PARAS = PARAS[~PARAS['para_str'].str.match(r'^s*$')] #remove empty paragraphs
```

```python
PARAS.head()
```

```python
#sent_pat = r'[.?!;:"]+' #paragraphs into sentence
sent_pat = r'[.?!;:]+'
SENTS = PARAS['para_str'].str.split(sent_pat, expand=True).stack()\
    .to_frame('sent_str')
SENTS.index.names = OHCO[:3]
```

```python
SENTS = SENTS[~SENTS['sent_str'].str.match(r'^\s*$')] #remove empty paragraphs
SENTS.sent_str = SENTS.sent_str.str.strip() #CRUCIAL TO RMEOVE BLANK TOKENS
```

```python
SENTS.head()
```

```python
#sentences into tokens
token_pat = r"[\s',-]+"
TOKENS = SENTS['sent_str'].str.split(token_pat, expand=True).stack()\
    .to_frame('token_str')
```

```python
TOKENS.index.names = OHCO[:4]
```

```python
TOKENS
```

```python
#extract vocab
TOKENS['term_str'] = TOKENS.token_str.replace(r'[\W_]+', '', regex=True).str.lower()
VOCAB = TOKENS.term_str.value_counts().to_frame('n').reset_index().rename(columns={'index':'term_str'})
```

```python
VOCAB
```

```python
#gathering by content object
def gather(ohco_level):
    global TOKENS
    level_name = OHCO[ohco_level-1].split('_')[0]
    df = TOKENS.groupby(OHCO[:ohco_level])\
        .token_str.apply(lambda x: x.str.cat(sep=' '))\
        .to_frame(f'{level_name}_str')
    return df
```

```python
gather(3)
```

```python
#save work to CSV
TOKENS.to_csv(csv_file)
```
